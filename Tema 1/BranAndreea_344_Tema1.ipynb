## 1. Incarcarea setului de date intr-o forma potrivita pentru PyTorch

Folosim `pandas` pentru a incarca `.csv-ul`. Setul de date trebuie separat in set de antrenare, de validare si de testare. Aceasta impartire nu se poate face complet aleator totusi, e necesar sa respectam distributia initiala a datelor in cele 3 subseturi. Pentru aceasta, putem folosi `StratifiedKFold` care imparte setul de date in k folduri de aceeasi distributie a claselor unei coloane alese. In cazul dat, vom respecta distributia numarului de camere ale apartamentelor.

Adaugam o coloana care va reprezenta fold-ul, doar pentru a putea separa detale.

```python
df = pd.read_csv("Bucharest_HousePriceDataset.csv")
display(df)
df["kfold"] = -1

# separam datele de intrare de etichete
X = df.drop('Nr Camere',axis=1).values
Y = df['Nr Camere'].values
```

```python
df['Nr Camere'].value_counts()
```

Uitandu-ne la matricea de mai sus, putem trage concluzia ca avem cateva apartamente care pot fi considerate valori extreme: un singur apartament cu 9 camere,2 cu 8 etc. Prin urmare, am decis sa refac categoriile astfel: categoria 5 va reprezenta orice apartament cu mai mult de 4 camere.

```python
df['Nr Camere'].clip(upper=5, inplace=True)
```

De asemenea, putem incerca sa mai adaugam un parametru care sa se apropie mai mult de numarul de camere. Intrucat pretul nu depinde doar de suprafata, ci si de numarul de camere, putem incerca sa adaugam un feature care sa reprezinte raportul acestora.

```python
df['feature_camere'] = df['Pret']/df['Suprafata']
```

Impartim datele astfel: 60% din date vor fi pentru antrenare, 20% pentru validare si 20% pentru test. Prin urmare, putem imparti in 5 folduri, dintre care setul de antrenare va lua 3. 
Cu cat antrenam pe mai multe date, cu atat putem avea o acuratete mai mare. In cazul de fata totusi, setul de date este destul de mic (~3500 de exemple), motiv pentru care daca alegem seturile de validare si cel de testare prea mici, rezultatele obtinute de model pe acestea ar putea sa nu fie suficient de relevante.

```python
sktf = StratifiedKFold(n_splits=5)
sktf.get_n_splits(X, Y)

# atribuim foldurile
for index,(x, l) in enumerate(sktf.split(X=df, y = Y)):
    df.loc[l, 'kfold'] = index

# separam cele 3 seturi si renuntam la coloana kfold
train = df[df.kfold<3]
train = train.drop('kfold', axis=1)

validation = df[df.kfold==3]
validation = validation.drop('kfold', axis=1)

test = df[df.kfold==4]
test = test.drop('kfold', axis=1)
```

Mai departe, vom vrea sa standardizam datele, in special deoarece coloana de `Pret` contine valori foarte mari. Pentru asta, vom folosi `MinMaxScaler` din `sklearn`

```python
scaler = MinMaxScaler()
scaler.fit(train.drop('Nr Camere',axis=1).values)
```

```python
display(train)
```

Cream o clasa care sa extinda clasa `Dataset` pentru a interpreta setul nostru de date. 

```python
class BucharestHousingDataset(Dataset):
  def __init__(self, dframe, transformer, target):
    self.full_dataset = dframe.values
    # transformer-ul va fi in cazul nostru un scaler
    self.transformer = transformer
    # setul fara nr. de camere
    self.X = self.transformer.transform(dframe.drop(target,axis=1).values)
    # coloana pe care o vom considera eticheta
    self.Y = dframe[target].values

  def __getitem__(self, index) -> torch.Tensor:
    # interpretarea fiecarei perechi (x,y) ca un dictionar in care doar x este tensor
    sample_as_tensor = torch.Tensor(self.X[index])
    return {"item": sample_as_tensor, "target": self.Y[index]}
  
  def __len__(self):
    return len(self.full_dataset)
```

Instatiem o astfel de clasa pentru train, validare si test. Toate aceste obiecte vor folosi drept transformer scalerul care a facut `.fit()` pe setul de train. Operatia de `fit` nu se face pe tot setul. Cand antrenam pe un set de date, noi presupunem ca acest este suficient de cuprinzator/relevant astfel incat sa putem prezice cat mai exact pe date ce nu au fost vazute in timpul antrenarii. De aceea, seturile de validare si test trebuie tratate cu obiectivitate, prin faptul ca fit-ul se face doar pe antrenare, si vor fi scalate conform mediei si variantei acestuia.

```python
train_dataset = BucharestHousingDataset(train, scaler, 'Nr Camere')
validation_dataset = BucharestHousingDataset(validation, scaler,'Nr Camere')
test_dataset = BucharestHousingDataset(test, scaler,'Nr Camere')
print("Dataset size: ", len(train_dataset))
print(train_dataset[2])
```

Folosim un sampler pentru a ne imparti setul de date de antrenare in batch-uri in mod aleator.

```python
train_sampler = RandomSampler(train_dataset)
train_dataloader = DataLoader(train_dataset, 
                        batch_size = 64, 
                        sampler = train_sampler)

validation_dataloader = DataLoader(validation_dataset, batch_size=64)

test_dataloader = DataLoader(test_dataset, batch_size=9)
```

# 2. Clasificare
Tratam problema ca pe o problema de clasificare. Definim un model simplu, cu doua straturi:
1. un strat hidden cu activare LeakyReLU, pentru a tine cont si de valorile negative
2. un strat de output cu activare softmax, pentru ca cele 9 valori obtinute sa fie transformate in probabilitati peste cele 9 clase

```python
class ClasificationModel(nn.Module):
    def __init__(self, 
                 input_size: int, 
                 hidden_size: int, 
                 output_size: int,
                 hidden_activation_fn = nn.LeakyReLU(),
                 classifier = nn.Softmax()):
        super().__init__()

        self._layer1 = nn.Linear(input_size, hidden_size)
        self._layer3 = nn.Linear(hidden_size, output_size)

        self._hidden_activation = hidden_activation_fn
        self._classifier = classifier
   
    def forward(self, x):
        h = self._layer1(x)
        h = self._hidden_activation(h)
        o = self._layer3(h)
        o = self._classifier(o)
        
        return o
```

Nu am ales un model cu multi parametri doarece avem foarte putine feature-uri (mai exact 6).
De asemenea, am folosit modelul de predictie si antrenare de la laborator pentru toate modelele, adaptat pentru prezicerea pe batch-uri.

*Functia de loss*: CrossEntropyLoss

```python
c_model = ClasificationModel(train.shape[1]-1, 16,5)

NUM_EPOCHS = 100
PRINT_EVERY = 10

# salvam valorile loss-urilor 
c_train_loss = []
c_val_loss = []

# definim optimizerul Adam si functia de loss CrossEntropyLoss
c_optimizer = torch.optim.Adam(c_model.parameters(), lr=3e-3)
c_criterion = nn.CrossEntropyLoss()


for epoch in range(NUM_EPOCHS):
    c_epoch_loss = 0.0
    for i, data_batch in enumerate(train_dataloader):
        # despachetam fiecare batch din dataloader
        x = data_batch["item"].float()
        y = data_batch["target"]-1
        
        c_optimizer.zero_grad()
        result= c_model(x)
        
        # calculam valoarea functiei de loss
        batch_loss = c_criterion(result, y)
        batch_loss.backward()
        c_optimizer.step()
        
        c_epoch_loss+=batch_loss.item()
    c_train_loss.append(c_epoch_loss/len(train))
    
    if epoch%10==0:
        print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, c_epoch_loss / len(train)))
    
    # folosim modelul de predictie din laborator, insa adaptat pentru a prezice pe batch-uri
    with torch.no_grad():
        validation_loss=0.0
        correct = 0
        for i, data_batch in enumerate(validation_dataloader):
            x = data_batch["item"].float()
            y = data_batch["target"]-1
            result= c_model(x)
            batch_loss = c_criterion(result, y)
            validation_loss += batch_loss.item()
            _, p_class = torch.max(result, 1) 
            correct += (p_class == y).sum().item()
            
        print('accuracy on validation: ', correct/len(validation))
        c_val_loss.append(validation_loss/len(validation))
    

print('Finished Training')
```

Afisam si procesul de minimizare al functiei de pierdere:

```python
plot_loss(c_train_loss, 'train-loss')
plot_loss(c_val_loss, 'val-loss', color='green')
```

Prezicem si pe setul de test:

```python
with torch.no_grad():
        correct = 0
        for i, data_batch in enumerate(test_dataloader):
            x = data_batch["item"].float()
            y = data_batch["target"]-1
            result= c_model(x)
            batch_loss = c_criterion(result, y)
            _, p_class = torch.max(result, 1) 
            correct += (p_class == y).sum().item()
        print('accuracy on test: ', correct/len(test))
```
